# Creating a basic text generation agent which will generate text in different ways

```python
from dotenv import load_dotenv
import os
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("API_KEY"))
```

```python
prompt = "tell me something about numbers"
```

```python
# This is the most basic way of generating a text

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant."
        },
        {
            "role": "user",
            "content": prompt
        }
    ],

    model="llama-3.3-70b-versatile"
)

# Print the completion returned by the LLM.
print(chat_completion.choices[0].message.content)
```

```python
# There are 2 types of parameters which we can pass
```

```python
stream = client.chat.completions.create(
    # Required parameters
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant."
        },
        {
            "role": "user",
            "content": "Explain the importance of fast language models",
        }
    ],

    model="llama-3.3-70b-versatile",

    # Optional parameters

    temperature=0.9, # temperature controls how creative or random the model's responses are.
    # Temperature	Use Case
    # 0 - 0.2	    Coding, math, factual answers
    # 0.3 - 0.7	    General chat, explanations
    # 0.8 - 1.0	    Creative writing, brainstorming
    # 1.0+	        Highly creative, unpredictable

    # The maximum number of tokens to generate. Requests can use up to
    max_completion_tokens=1024,

    top_p=1,# It cuts off low-probability options.
    # top_p = 0.5 For this only the first word is needed to reach 50%.
    # top_p = 0.8 for this all the words will be selected until cumulative probability reaches 80%
    # top_p = 1 for this all the words will be selected 
    

    # A stop sequence is a predefined or user-specified text string that
    # signals an AI to stop generating content, ensuring its responses
    # remain focused and concise. Examples include punctuation marks and
    # markers like "[end]".
    stop=None,

    # If set, partial message deltas will be sent.
    stream=True, # enables streaming responses, meaning the model sends tokens as they are generated instead of waiting for the 
    # complete response. Without this the model will wait until the response it fully completed
)

# Print the incremental deltas returned by the LLM.
for chunk in stream:
    print(chunk.choices[0].delta.content, end="")
```

Stop sequences allow you to control where the model should stop generating. When the model encounters any of the specified stop sequences, it will halt generation at that point. This is useful when you need responses to end at specific points.

```python
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant."
        },
        {
            "role": "user",
            "content": "Count to 10.  Your response must begin with \"1, \".  example: 1, 2, 3, ...",
        }
    ],
    model="llama-3.3-70b-versatile",

    temperature=0.5,

    max_completion_tokens=1024,


    top_p=1,

    # A stop sequence is a predefined or user-specified text string that
    # signals an AI to stop generating content, ensuring its responses
    # remain focused and concise. Examples include punctuation marks and
    # markers like "[end]".
    # For this example, we will use ", 6" so that the llm stops counting at 5.
    # If multiple stop values are needed, an array of string may be passed,
    # stop=[", 6", ", six", ", Six"]
    stop="6",

    # If set, partial message deltas will be sent.
    stream=False,
)

# Print the completion returned by the LLM.
print(chat_completion.choices[0].message.content)
```
