# Prompt Engineering

```python
We can broadly say that writing good prompts is based on 3 principles:

1. Write clear and specific instructions
2. Choose the optimal model based on your requirement
3. Defining negative edge cases is more important than defining positive edge cases
```

```python
We should use these tactics with the above-mentioned principles to get a desired output:

1. Use delimiters everywhere to structure your input
2. Ask for a structured output
3. Check whether conditions are satisfied; also check for the assumptions required to complete a task
4. Give an example of successfully completing the task
5. Specify the steps to complete a task
```
