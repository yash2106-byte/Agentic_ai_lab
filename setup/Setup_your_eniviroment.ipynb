```python
from dotenv import load_dotenv
import os
from groq import Groq

load_dotenv()
client = Groq(api_key=os.getenv("API_KEY"))
```

```python
# client is your Groq API client.
# chat accesses the chat API.
# completions is the endpoint that generates responses.
# create() sends a request to Groq's servers.
we can assign different roles which will help us to get our desired response
# Role	Purpose
# system	Instructions for the AI
# user	Message from the user
# assistant	Previous AI responses
```

```python
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": "Explain the importance of fast language models",
        }
    ],
    model="llama-3.3-70b-versatile",
)

print(chat_completion.choices[0].message.content)
```

```python
# This is how we can set different roles in the same querry
chat_completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
            "role": "system",
            "content": "Give the output in less than 50 words."
        },
        {
            "role": "user",
            "content": "Explain the importance of fast language models."
        }
    ]
)

print(chat_completion.choices[0].message.content)
```

```python
# we get a complete respone in the form of an object 
print("this is the object which is returend")
print(chat_completion)
print("this is the message which we get as a respone")
print(chat_completion.choices[0].message)
print("this is the role which we get as a respone")
print(chat_completion.choices[0].message.role)
```

```python
# Hence we have to format it 
.choices[0] = The API can generate multiple responses hence we only print the first one
.message.content = Inside the first choice we have role and content so we take only content from there
```

```python
def get_completion(prompt, model="llama-3.1-8b-instant"):
    response = client.chat.completions.create(
        messages=[{"role": "user", "content": prompt}],
        model=model,
        temperature=0,
    )
    return response.choices[0].message.content
```

```python
print(get_completion("Say hello in one word."))
```
